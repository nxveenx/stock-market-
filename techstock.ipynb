```python
import pandas_datareader.data as web
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline
```

```python
ticker_symbol = ['TCS']
start_date= '2022-11-01'
end_date='2023-11-01'
tcs= yf.download(ticker_symbol,start=start_date,end=end_date)
tcs
```

```python
ticker_symbol = ['infy']
start_date= '2022-11-01'
end_date='2023-11-01'
infy= yf.download(ticker_symbol,start=start_date,end=end_date)
infy
```

```python
tcs["Open"].plot(label='TCS',figsize=(15,7))
infy["Open"].plot(label='infosys')
plt.legend()
plt.title("Tcs and Infosys opening prices")
plt.ylabel("price")
```

```python
tcs['Volume'].plot(label='tcs',figsize=(15,7))
infy['Volume'].plot(label='infosys')
plt.legend()
plt.title("volumes sold by tcs and Infosys")
plt.ylabel("price")
```

total traded

```python
tcs['Total traded']=tcs['Open']*tcs['Volume']
infy['Total traded']=infy['Open']*infy['Volume']
tcs.head()
```

```python
infy.head()
```

```python
tcs['Total traded'].plot(label="tcs",figsize=(15,7))
infy['Total traded'].plot(label='infosys')
plt.legend()
plt.title("Total volumes traded")
plt.ylabel("price")
```

Moving average

```python
infy["Open"].plot(figsize=(15,7))
infy['MA 50']=infy['Open'].rolling(50).mean()
infy['MA 50'].plot(label='MA 50')
infy['MA 20']=infy['Open'].rolling(20).mean()
infy['MA 20'].plot(label='MA 30')
plt.legend()
```

```python
tcs["Open"].plot(figsize=(15,7))
tcs['MA 50']=tcs['Open'].rolling(50).mean()
tcs['MA 50'].plot()
plt.legend()
```
